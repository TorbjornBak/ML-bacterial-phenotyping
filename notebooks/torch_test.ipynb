```python
def compress_integer_embeddings(integer_embeddings, alphabet_size, kmer_suffix_size):
	# Input: dict of the embeddings
	
	arr1 = np.zeros(4**kmer_suffix_size + 1, dtype = np.uint64)
	# Detect which embeddings are used, which aren't
	for embeddings_np in integer_embeddings:
		unique_embeddings = np.unique(embeddings_np)
		for embedding in unique_embeddings:
			#print(embedding)
			arr1[embedding] = embedding
	
	arr2 = [np.uint64(0)]
	arr2.extend([a for a in arr1[1:] if a != 0])

	mapping = {old_embedding:new_embedding for new_embedding, old_embedding in enumerate(arr2)}
	
	
	print(len(mapping))
	print(mapping)

	compressed_embeddings = dict()
	for i, embeddings_np in enumerate(integer_embeddings):
		compressed_embeddings[i] = np.array([mapping[emb] for emb in embeddings_np])
		

	return arr1, arr2, compressed_embeddings
```
